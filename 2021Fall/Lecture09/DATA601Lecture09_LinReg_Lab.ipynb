### Advertising Dataset
- The Advertising dataset consists of the sales of that product in 200 different markets, along with advertising budgets for the product in each of those markets for three different media: TV, radio, and newspaper.
- It is not possible for our client to directly increase sales of the product, but they can control the advertising expenditure in each of the three media.
- Therefore, if we determine that there is an association between advertising and sales, then we can instruct our client to adjust advertising budgets, thereby indirectly increasing sales.

```python
# get the data
advertising = pd.read_csv('https://raw.githubusercontent.com/simsekergun/DATA601/main/2021Fall/datasets/Advertising.csv', index_col=0)
```

### Step 1
 <b>1.1</b> Build a linear regression model to estimate the Sales using the TV advertisement. <br>
 Use 80% of the dataset for training and 20% for the testing, and calculate RMSE and R$^2$. <br>
 

```python
advertising.head()
```

```python
X = advertising['TV']
Y = advertising['Sales']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
```

```python
pd.DataFrame(X_train)
```

```python
##  Linear Regression Fitting example
model = LinearRegression()
model.fit(pd.DataFrame(X_train), pd.DataFrame(Y_train))
#
## prediction example
y_test_predict = model.predict(pd.DataFrame(X_test))
#
## model evaluation
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)
print(rmse,r2)
```

```python
advertising.describe()
```

```python
plt.scatter(X_test.values, Y_test.values,color='b')
plt.scatter(X_test.values, y_test_predict,color='r')
plt.show()
```

```python
X = advertising['Radio']
Y = advertising['Sales']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)

##  Linear Regression Fitting example
model = LinearRegression()
model.fit(pd.DataFrame(X_train), pd.DataFrame(Y_train))
#
## prediction example
y_test_predict = model.predict(pd.DataFrame(X_test))
#
## model evaluation
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)
print(rmse,r2)
```

```python
X = advertising['Newspaper']
Y = advertising['Sales']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)

##  Linear Regression Fitting example
model = LinearRegression()
model.fit(pd.DataFrame(X_train), pd.DataFrame(Y_train))
#
## prediction example
y_test_predict = model.predict(pd.DataFrame(X_test))
#
## model evaluation
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)
print(rmse,r2)
```

```python
X = advertising[['TV', 'Radio','Newspaper']]
Y = advertising['Sales']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)

##  Linear Regression Fitting example
model = LinearRegression()
model.fit(pd.DataFrame(X_train), pd.DataFrame(Y_train))
#
## prediction example
y_test_predict = model.predict(pd.DataFrame(X_test))
#
## model evaluation
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)
print(rmse,r2)
```
